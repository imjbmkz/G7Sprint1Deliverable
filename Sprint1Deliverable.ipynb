## Data Cleaning and Preprocessing

```python
df_copy = df.copy()
```

```python
df_copy['dob'] = pd.to_datetime(df_copy['dob'], format='%d/%m/%Y')
```

```python
display(df_copy)
```

```python
df_copy['city_pop'] = df_copy['city_pop'].str.replace('people', '').str.replace(',', '')
```

```python
df_copy['city_pop'] = pd.to_numeric(df_copy['city_pop'])
```

```python
display(df_copy['city_pop'])
```

```python
df_copy.info()
```

```python
display(df_copy['gender'])
```

```python
df_copy['gender'].unique()
```

```python
df_copy['gender'] = df_copy['gender'].replace({'M' : 'Male', 'F' : 'Female'})
```

```python
display(df_copy)
```

```python
df_copy['gender'].unique()
```

```python
df_copy['trans_timestamp'] = pd.to_datetime(df_copy['unix_time'], unit='s')
display(df_copy['trans_timestamp'])
```

```python
df_copy.dtypes
```

```python
df_copy['trans_timestamp'].min(), df_copy['trans_timestamp'].max()
```

```python
# Formulating a new trans_date column that retrieves the date only.
df_copy['trans_date'] = pd.to_datetime(df_copy['trans_timestamp'].dt.strftime('%Y-%m-%d'), format='%Y-%m-%d')
display(df_copy['trans_date'])
```

```python
df_copy['amt'] = df_copy['amt'].replace('\$', '', regex=True)
display(df_copy['amt'])
```

```python
df_copy['amt'] = pd.to_numeric(df_copy['amt'])
display(df_copy['amt'])
```

```python
df_copy.info()
```

```python
df_copy[(df_copy['acct_num'] != df_copy['acct_num2'])]
```

```python
df_copy = df_copy.drop(['acct_num2'], axis=1)
```

```python
df_copy.columns
```

```python
sorted(df_copy['city'].unique())
```

## EDA

How much is the total amount of transaction per account holder?

```python
df_copy.groupby('acct_num')['amt'].sum()
```

```python
df_copy.pivot_table(index='acct_num', aggfunc={'amt' :'sum'})
```

How many males are there compared to females?

```python
df_copy['gender'].value_counts()
```

```python
df_copy.groupby('gender')['gender'].count()
```

What is the gender-specific spending behavior within each product category, and how does it vary in terms of the total spending, average spending, spending distribution (median), spending variability (standard deviation), and spending consistency (variance)?

```python
df_copy.groupby(['category', 'gender'])['amt'].agg(['sum', 'mean', 'median', 'std', 'var'])
```

```python
df_copy['category'].isna().sum() # We still see there are missing categories
```

What are the spending patterns among customers with unspecified product categories, categorized by gender, in terms of their total spending, average spending, and median spending?

```python
df_copy[df_copy['category'].isna()].groupby(['gender'])['amt'].agg(['sum', 'mean', 'median'])
```

```python
df_copy[df_copy['category'].isna()][['category', 'amt']].describe()
```

```python
df_copy['category'].value_counts()[:5] # Getting the top 5 most frequent category
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_copy.groupby('trans_date')['amt'].mean()
```

```python
# Sorting the trans_date column for plotting
df_copy.sort_values(by='trans_date', inplace=True)
display(df_copy['trans_date'])
```

```python
# Reseting the index after sorting
df_copy.reset_index(drop=True, inplace=True)
```

```python
# Plotting the amount vs transaction date from 2020-01 till the end
plt.plot(df_copy['trans_date'], df_copy['amt'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Amount Over Years')
```

```python
# Plotting only the year 2021 and for the months August to December
filtered_date = df_copy[(df_copy['trans_date'].dt.year == 2021) & (df_copy['trans_date'].dt.month >= 8)]

plt.plot(filtered_date['trans_date'], filtered_date['amt'], linestyle='-')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Amount Over Years')
```

```python
# Getting the sum of transactions of the year 2021
df_copy[df_copy['trans_date'].dt.year == 2021].groupby(df_copy['trans_date'].dt.month)['amt'].sum()
```

```python
# Plotting the Category vs Amount on a horizontal bar plot
df_copy.groupby('category')['amt'].agg('mean').plot(kind='barh')
```
